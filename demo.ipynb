```python
import numpy as np
import matplotlib.pyplot as plt
import os

from pymaze import maze as mg
from pymaze import astar as AStar
```

```python
maze_generator = mg.MazeGenerator([51, 51])
maze_generator.generate(save = False)
maze_generator.check_maze()
```

```python
field = maze_generator.get_maze()
start, goal = maze_generator.get_points()["start"], maze_generator.get_points()["goal"]
print(start, goal)
```

```python
astar = AStar.AStar(field=field, start = start, goal = goal)
astar.set_cost_function("c")
```

```python
astar.search(save=False)
```
